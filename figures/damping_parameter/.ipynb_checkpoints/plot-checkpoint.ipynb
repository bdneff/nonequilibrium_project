```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import os

# === Define model: damped harmonic oscillator with variable amplitude ===
def damped_oscillator_zeta(t, A, zeta, nu0):
    exp_term = np.exp(-2 * np.pi * zeta * nu0 * t)
    omega_d = 2 * np.pi * nu0 * np.sqrt(1 - zeta**2)
    cos_term = np.sqrt(1 - zeta**2) * np.cos(omega_d * t)
    sin_term = zeta * np.sin(omega_d * t)
    return A * exp_term * (cos_term - sin_term)

# === Time grid ===
t_ps = np.linspace(-2, 2, 1000)
t_pos = t_ps[t_ps > 0]

# === THz to cm^-1 conversion ===
THz_to_cm = 33.35641

# === Data paths for each frequency slice ===
paths = {
    'freq1': '/Users/bdneff/Downloads/non_equil_project/1d_vdos/equil/freq1_1d_vdos.dat',
    'freq11': '/Users/bdneff/Downloads/non_equil_project/1d_vdos/equil/freq12_1d_vdos.dat',
    'freq23': '/Users/bdneff/Downloads/non_equil_project/1d_vdos/equil/freq24_1d_vdos.dat',
}

# === Label & color mappings for plot ===
colors = {
    'freq1': '#8B0000',       # dark red
    'freq11': '#1E90FF',      # dodger blue
    'freq23': '#FF1493',      # deep pink
}
labels = {
    'freq1': '≈ 0 cm⁻¹',
    'freq11': '≈ 100 cm⁻¹',
    'freq23': '≈ 200 cm⁻¹',
}

# === Process all data and extract zeta values ===
results = {}
for label, path in paths.items():
    if not os.path.exists(path):
        print(f"File not found: {path}")
        continue

    data = np.loadtxt(path)
    freqs_cm = data[0]
    omega = 2 * np.pi * 2.99792458e10 * freqs_cm * 1e-12  # angular frequency in rad/ps
    vdos_modes = data[1:]

    zetas = []
    for vdos in vdos_modes:
        vacf = np.array([
            np.trapz(vdos / 1e18 * np.cos(omega * t), x=omega)
            for t in t_ps
        ])
        vacf_pos = vacf[t_ps > 0]

        try:
            popt, _ = curve_fit(
                damped_oscillator_zeta,
                t_pos,
                vacf_pos,
                p0=[1.0, 0.05, 5],  # A, ζ, ν₀
                bounds=([0.01, 0.001, 0.1], [10.0, 0.9, 50])
            )
            _, zeta_fit, _ = popt
            zetas.append(zeta_fit)
        except RuntimeError:
            zetas.append(np.nan)

    results[label] = np.array(zetas)
```

```python
# === Plotting ζ vs. mode index ===
plt.figure(figsize=(16, 8))
for label in ['freq1', 'freq11', 'freq23']:
    zeta_vals = results.get(label, [])
    valid = ~np.isnan(zeta_vals)
    plt.plot(
        np.arange(len(zeta_vals))[valid][:200],
        zeta_vals[valid][:200],
        label=labels[label],
        color=colors[label],
        linewidth=3
    )

plt.xlabel("Mode Index", fontsize=24)
plt.ylabel("Damping Ratio ζ", fontsize=24)
plt.xlim(-10, 210)
plt.ylim(-.1, 1.1)
plt.xticks(fontsize=24)
plt.yticks(fontsize=24)
plt.legend(fontsize=16, loc="lower right")
plt.tight_layout()

# Save and show
plt.savefig("Damping_vs_ModeIndex.png", dpi=300)
plt.show()
```
